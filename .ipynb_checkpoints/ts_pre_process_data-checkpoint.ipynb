```python
import os
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from matplotlib import pyplot
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('df_germany.csv', converters={'ags2': str, 'ags5': str})
tmp = pd.read_csv('oeffentliche_finanzen.csv', converters={'ags2': str, 'ags5': str} )
df = pd.merge(df, tmp[['ags2','ags5','kreis']])
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
df = df[df['kreis'] == 'Berlin']
df = df[['date','unemployment_rate','number_of_company_deletions','number_of_start_ups']]
df = df.dropna()
df = df.rename(columns={'number_of_company_deletions': 'number_of_companies_closed'})
df = df[df['date'] < '2019-01-01']
df = df.reset_index(drop=True)
df
```
